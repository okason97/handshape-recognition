```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Ciarp

axes = plt.axes()
# axes.set_ylim([80, 105])
axes.set_xticks(ticks=[44, 67, 85, 100])

df = pd.DataFrame({
    'x': [44, 67, 85, 100], 
    'proto-net': [99.97, 100.0, 99.99, 99.99],
    'proto-net++': [99.99, 100.0, 100.0, 100.0],
    'dense-net': [99.76, 99.84, 99.90, 99.93],
    'dense-net++': [99.86, 99.90, 99.92, 99.99]
})

plt.xlabel('samples percentage')
plt.ylabel('accuracy')
    
plt.plot('x', 'proto-net', label='Prototypical Networks', data=df, marker='o', markerfacecolor='blue', markersize=2, color='skyblue', linewidth=1)
plt.plot('x', 'proto-net++', label='Prototypical Networks ++', data=df, marker='o', markerfacecolor='blue', markersize=2, color='skyblue', linewidth=1, linestyle='dashed')
plt.plot('x', 'dense-net', label='DenseNet', data=df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=1)
plt.plot('x', 'dense-net++', label='DenseNet ++', data=df, marker='o', markerfacecolor='red', markersize=2, color='red', linewidth=1, linestyle='dashed')

plt.legend()

plt.savefig('ciarp.png')
```
